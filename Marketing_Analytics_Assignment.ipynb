First, let's read in the three CSV files again, as the previous loading cells were deleted. We'll use `pandas` to load `01_facebook_ads.csv`, `02_google_ads.csv`, and `03_tiktok_ads.csv` into separate DataFrames.

```python
import pandas as pd

# Read 01_facebook_ads.csv as 'facebook'
csv_file_path_facebook = '/content/01_facebook_ads.csv'
try:
    facebook_df = pd.read_csv(csv_file_path_facebook)
    print(f"Successfully read {csv_file_path_facebook}")
except FileNotFoundError:
    print(f"Error: The file {csv_file_path_facebook} was not found.")
    facebook_df = None

# Read 02_google_ads.csv as 'google'
csv_file_path_google = '/content/02_google_ads.csv'
try:
    google_df = pd.read_csv(csv_file_path_google)
    print(f"Successfully read {csv_file_path_google}")
except FileNotFoundError:
    print(f"Error: The file {csv_file_path_google} was not found.")
    google_df = None

# Read 03_tiktok_ads.csv as 'tiktok'
csv_file_path_tiktok = '/content/03_tiktok_ads.csv'
try:
    tiktok_df = pd.read_csv(csv_file_path_tiktok)
    print(f"Successfully read {csv_file_path_tiktok}")
except FileNotFoundError:
    print(f"Error: The file {csv_file_path_tiktok} was not found.")
    tiktok_df = None

print("\nInitial DataFrames loaded.")
```

Now, let's normalize the column names across the three DataFrames to ensure consistency for combined analysis. We'll rename columns with similar meanings to a consistent name, such as `ad_group_id`, `ad_group_name`, `spend`, and `views`.

```python
# Normalize columns for facebook_df
if facebook_df is not None:
    facebook_df = facebook_df.rename(columns={
        'ad_set_id': 'ad_group_id',
        'ad_set_name': 'ad_group_name',
        'video_views': 'views', # Align with TikTok's 'views'
        'cost': 'spend'
    })
    facebook_df['platform'] = 'Facebook'
    print("Facebook DataFrame columns normalized and 'platform' column added. First 5 rows:")
    display(facebook_df.head())

# Normalize columns for google_df
if google_df is not None:
    google_df = google_df.rename(columns={
        'cost': 'spend'
    })
    google_df['platform'] = 'Google'
    print("\nGoogle DataFrame columns normalized and 'platform' column added. First 5 rows:")
    display(google_df.head())

# Normalize columns for tiktok_df
if tiktok_df is not None:
    tiktok_df = tiktok_df.rename(columns={
        'adgroup_id': 'ad_group_id',
        'adgroup_name': 'ad_group_name'
        # 'views' is already consistent
    })
    tiktok_df['platform'] = 'TikTok'
    print("\nTikTok DataFrame columns normalized and 'platform' column added. First 5 rows:")
    display(tiktok_df.head())
```

Finally, we'll combine these three normalized DataFrames into a single, comprehensive dataset. We will create a `verified_df` by concatenating `facebook_df`, `google_df`, and `tiktok_df`.

```python
# Combine the three DataFrames into a single DataFrame
combined_dfs = []
if facebook_df is not None: combined_dfs.append(facebook_df)
if google_df is not None: combined_dfs.append(google_df)
if tiktok_df is not None: combined_dfs.append(tiktok_df)

if combined_dfs:
    verified_df = pd.concat(combined_dfs, ignore_index=True)
    print("Combined DataFrame created. First 5 rows:")
    display(verified_df.head())
    print("\nCombined DataFrame Info:")
    display(verified_df.info())
else:
    verified_df = pd.DataFrame()
    print("No DataFrames were available to combine.")
```

Now, let's select only the relevant columns up to 'conversions' and update the 'platform' column based on the `campaign_id` prefix.

```python
import numpy as np

# Select columns up to and including 'conversions'
# Note: Assuming 'platform' is kept if it exists, as it's modified next.
# If 'platform' should be dropped and re-added from scratch based on new logic,
# the column selection would need to be adjusted.
selected_columns = [
    'date', 'campaign_id', 'campaign_name', 'ad_group_id', 'ad_group_name',
    'impressions', 'clicks', 'spend', 'conversions'
]

# Ensure 'platform' column is in the selected_columns for modification if it's not already there
# In this case, 'platform' is already present due to previous steps, and we want to redefine its content
# so we re-create it.

# Create a new DataFrame with the specified columns
filtered_df = verified_df[selected_columns].copy()

# Add/Update the 'platform' column based on campaign_id
conditions = [
    filtered_df['campaign_id'].str.startswith('fb_'),
    filtered_df['campaign_id'].str.startswith('g_'),
    filtered_df['campaign_id'].str.startswith('tt_')
]
choices = ['fb', 'g', 'tt']
filtered_df['platform'] = np.select(conditions, choices, default='other')

print("DataFrame filtered to selected columns and 'platform' column updated. First 5 rows:")
display(filtered_df)

print("\nFiltered DataFrame Info:")
display(filtered_df.info())
```

```python
import plotly.io as pio
pio.renderers.default = "colab" # or "png"
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# 1. Prepare Aggregated Data
# Aggregate daily for trends
daily_metrics = filtered_df.groupby(['date', 'platform']).agg({
    'spend': 'sum',
    'conversions': 'sum'
}).reset_index()

# Aggregate by platform for total metrics
platform_summary = filtered_df.groupby('platform').agg({
    'spend': 'sum',
    'impressions': 'sum',
    'clicks': 'sum',
    'conversions': 'sum'
}).reset_index()

# Calculate Ratios
platform_summary['CTR'] = (platform_summary['clicks'] / platform_summary['impressions']) * 100
platform_summary['CPA'] = platform_summary['spend'] / platform_summary['conversions']

# 2. Create the Dashboard
# Row 1: Totals
fig1 = make_subplots(
    rows=1, cols=4,
    subplot_titles=('Total Spend', 'Total Impressions', 'Total Clicks', 'Total Conversions')
)

metrics = ['spend', 'impressions', 'clicks', 'conversions']
for i, metric in enumerate(metrics):
    fig1.add_trace(go.Bar(x=platform_summary['platform'], y=platform_summary[metric], name=metric, marker_color='royalblue'), row=1, col=i+1)

fig1.update_layout(height=400, title_text='Aggregate Volume Metrics by Platform', showlegend=False)
fig1.show()

# Row 2: Trends
fig2 = px.line(daily_metrics, x='date', y='spend', color='platform', title='Daily Spend Trend by Platform')
fig2.update_layout(height=400)
fig2.show()

fig3 = px.line(daily_metrics, x='date', y='conversions', color='platform', title='Daily Conversions Trend by Platform')
fig3.update_layout(height=400)
fig3.show()

# Row 3: Efficiency Ratios
fig4 = make_subplots(
    rows=1, cols=2,
    subplot_titles=('CTR (clicks/impressions %)', 'CPA (cost/conversions $)')
)

fig4.add_trace(go.Bar(x=platform_summary['platform'], y=platform_summary['CTR'], name='CTR', marker_color='orange'), row=1, col=1)
fig4.add_trace(go.Bar(x=platform_summary['platform'], y=platform_summary['CPA'], name='CPA', marker_color='indianred'), row=1, col=2)

fig4.update_layout(height=400, title_text='Efficiency Metrics: CTR & CPA', showlegend=False)
fig4.show()
```
